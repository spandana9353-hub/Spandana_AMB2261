```python
import numpy as np
arr=np.array([[1,2,3],[4,5,6]])
print(arr)
print("originalarr.shape:",arr.shape)
print()
transposed_arr=arr.transpose()
print(transposed_arr)
print("transposed_arr.shape:",transposed_arr.shape)
print()
flattened_arr=arr.flatten()
print(flattened_arr)
print("flattened_arr.shape:",flattened_arr.shape)
```
