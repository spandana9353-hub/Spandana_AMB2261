```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
data_set =pd.read_csv('Salary_Data (1).csv')
data_set.head()
```

```python
# slice dataset as descriptive(independent)
#target variable (dependent)
x=data_set.iloc[:,0:1].values
y=data_set.iloc[:,-1].values
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=1/3,random_state=0)
```

```python
print(data_set.shape)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# calculate the mean values of input and output
x_mean=np.mean(x_train)
y_mean=np.mean(y_train)
print(x_mean,y_mean)
```

```python
# calculate intercept and cofficent
n=len(x_train)
num,den=0,0
for i in range (n):
  num+=(x_train[i]-x_mean)*(y_train[i]-y_mean)
  den+=(x_train[i]-x_mean)**2
b1=num/den
b0=y_mean-(b1*x_mean)
print("b0:",b0)
print("b1:",b1)
```

```python
# predicting results on test dataset
y_pred_code=b0+b1*x_test
print(y_pred_code)
```

```python
print(y_test)
```

```python
# mean square error
n1=len(y_test)
mse=0
for i in range (n1):
 mse+=(y_test[i]-y_pred_code[i])**2
mse=mse/n1
print(mse)
```

```python
# root mean square error
rmse=np.sqrt(mse)
print(rmse)
```

MECHINE LEARNING ALGORITHM
linear regression using inbuilt function

```python
# create an object of the algorithm /model
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
#train the model
reg.fit(x_train,y_train)
```

```python
# predict the model on testing dataset
y_pred=reg.predict(x_test)
print(y_pred)
```

```python
# calculate the intercept and coefficent
print("intercept",reg.intercept_)
print("coefficent",reg.coef_)
```

```python
# evaluation metrics
from sklearn import metrics
print("mean absolute error:")
print(metrics.mean_absolute_error(y_test,y_pred))
print("mean squared error:",metrics.mean_squared_error(y_test,y_pred))
print("root mean squared error:",metrics.root_mean_squared_error(y_test,y_pred))
```

```python
# Visualization
plt.scatter(x_train,y_train,color="green")
plt.plot(x_train,reg.predict(x_train,),color="blue")
plt.title("salary vs experience(training set)")
plt.xlabel("years of experience")
plt.ylabel("salary")
plt.show()
```
