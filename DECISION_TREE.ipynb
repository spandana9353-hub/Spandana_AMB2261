```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
dataset=pd.read_csv("/content/iris (1).csv")
print(dataset)
```

```python
dataset["Species"].unique()
```

```python
dataset["Species"].value_counts()
```

```python
#seperate dependent and independent variables (input/output)
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,4].values
```

```python
#split the dataset into train and test model
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=1/3,random_state=0)
```

```python
# create decision tree classifer object
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(criterion="entropy",max_depth=None,min_samples_split=5,min_samples_leaf=2)
```

```python
#TRAIN THE MODEL(fit method)
model.fit(x_train,y_train)
```

```python
#predict the testing model
y_pred=model.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix
c=confusion_matrix(y_test,y_pred)
print(c)
```

```python
len(y_test)
```

```python
from sklearn.metrics import accuracy_score
c=accuracy_score(y_test,y_pred)
c
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn import tree
fn=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn=['setosa','versicolor','virginica']
fig,axes=plt.subplots(nrows=1,ncols=1,figsize=(4,4),dpi=200)
tree.plot_tree(model,feature_names=fn,class_names=cn,filled=True);
```
