# 1: Data Structure - Tuple

### 1.1: Tuple declaration

```python
# Method 1

T = ("Pratham", 'Sharma', 3.14, 3)

print ("T                  -->", T)
print ("Number of elements -->", len(T))
print ("Type of Object     -->", type(T))
```

```python
# Method 2

T = tuple(["Pratham", 'Sharma', 3.14, 3])   # Convert list to tuple
#T = tuple(("Pratham", 'Sharma', 3.14, 3))  # Also Works

print ("T                  -->", T)
print ("Number of elements -->", len(T))
print ("Type of Object     -->", type(T))
```

### 1.2: Tuple with one item

```python
# Example 1: Tuple with one item

T = ("Pratham",)            # atleast one comma is must

print ("T                  -->", T)
print ("Number of elements -->", len(T))
print ("Type of Object     -->", type(T))
```

```python
# Example 2: This is not a Tuple. It is a string

T = ("Pratham")         # Without comma, it is a string

print ("T                  -->", T)
print ("Number of elements -->", len(T))
print ("Type of Object     -->", type(T))
```

### 1.3: Tuple iteration using while loop

```python

T = ("Pratham", 'Sharma', 3.14, 3)
print ("T -->", T)

i = 0
print ("")
while i < len(T):
    print (T[i])
    i += 1
```

### 1.4: Tuple iteration using for loop

```python
# Example 1:

T = ("Pratham", 'Sharma', 3.14, 3)
print ("T -->", T)

print ("")
for i in range( len(T)):   # refereing to the indexes
    print (T[i])
```

```python
# Example 2:

T = ("Pratham", 'Sharma', 3.14, 3)
print ("T -->", T)

print ("")
for s in T:         #refereing to the elements
    print (s)
```

# 2: Data Structure - Sets

### 2.1: Sets declaration

```python
# Example 1

#list   []
#tuple ()
#sets {}
#dictionary {k:v}

S = {'E','A', 'B', 'E', 'F','E', 'F'}

print ("S                  -->", S)
print ("Number of elements -->", len(S))
print ("Type of Object     -->", type(S))
```

```python
# Example 2

S = set(['A', 'B', 'E', 'F','E', 'F' ])
# S = set(('A', 'B', 'E', 'F','E', 'F' )) # Also works

print ("S                  -->", S)
print ("Number of elements -->", len(S))
print ("Type of Object     -->", type(S))
```

### 2.2: Traverse set elements

```python

S = {'A', 'B', 'E', 'F','E', 'F'}

print ("S -->", S)

print ("")
for k in S:
    print (k)
```

### 2.3: Set elements are unindexed

```python

S = {'A', 'B', 'E', 'F','E', 'F'}

print ("S -->", S)

print (S[2])   # Error; 'set' object is not subscriptable
```

### 2.4: Adding element to Set using add()

```python

S = set(['A', 'B', 'E', 'F' ])
print ("S                -->", S)

S.add("D")
print ("After Adding (D) -->", S)

S.add("D")
print ("After Adding (D) -->", S)    # No Change in Set

S.add("C")
print ("After Adding (C) -->", S)
```

### 2.5: Adding element to Set using update()

```python

# Once a set is created, we cannot change the items, but we can add new items.

#Example 1:
S1 = set(['A', 'B', 'E', 'F' ])
S2 = set(["D", "E", "F", "G" ])

print ("S1              -->", S1)
print ("S2              -->", S2)

S1.update(S2)                       #inserts the items in S2 into S1; Union

print ("\nAfter update S1 -->", S1)
```

```python
#Example 2:

S = set(['A', 'B', 'E', 'F' ])
print ("S              -->", S)

S.update(["F", "G", "H", "I"])

print ("After update S -->", S)
```

### 2.6: Removing element from Set using remove()

```python

S = set(['A', 'B', 'E', 'F' ])
print ("S                  -->", S)

S.remove("B")
print ("After removing (B) -->", S)

S.remove("E")
print ("After removing (E) -->", S)

S.remove("E")                           # Error
print ("After Adding (E) -->", S)

#Note: If the item does not exist, remove() will raise an exception
```

### 2.7: Removing element from Set using discard()

```python

S = set(['A', 'B', 'E', 'F' ])
print ("S                 -->", S)

S.discard("B")
print ("After discard (B) -->", S)

S.discard("E")
print ("After discard (E) -->", S)

S.discard("E")                         # No Error
print ("After discard (E) -->", S)

#Note: If the item does not exist, discard() will NOT raise an exception
```

### 2.8: Removing element from Set using pop()

```python

S = set(['A', 'B' ])
print ("S         -->", S)

S.pop()
print ("After pop -->", S)

S.pop()
print ("After pop -->", S)

S.pop()                       # Error; Set is empty; Under flow
print ("After pop -->", S)


#Note: If set is empty, pop() will raise an exception
```

### 2.14: Operations on Sets - Union, Intersection, Difference, Symmetric  Difference

```python

a = {'A', 'B', 'E', 'F'}
b = set(["A", "C", "D", "E"])

print ("Set a -->", a)
print ("Set b -->", b)

print("")
print ("Union of a and b          -->", a.union(b))    #ABEFCD
print ("Intersection of a and b   -->", a.intersection(b))  #AE

print("")
print ("Difference a - b          -->", a - b)
print ("Difference a - b          -->", a.difference(b))    #BF
print ("Difference b - a          -->", b - a)
print ("Difference b - a          -->", b.difference(a))     #CD

print("")
print ("Symetric Difference a - b -->", a.symmetric_difference(b))
print ("Symetric Difference b - a -->", b.symmetric_difference(a))
```

```python
In Python, the mutability or immutability of a data type refers to
whether the object's state can be modified after it is created. '
'Here's a brief overview of common data types and their mutability:

Mutable:

List (list): Lists are mutable. You can modify, add, or remove elements after the list is created.
Dictionary (dict): Dictionaries are mutable. You can add, remove, or update key-value pairs after the dictionary is created.
Set (set): Sets are mutable. You can add and remove elements from a set after it is created.
Immutable:

Tuple (tuple): Tuples are immutable. Once a tuple is created, you cannot modify its elements.
String (str): Strings are immutable. You cannot change the characters in a string after it is created.
Integer (int): Integers are immutable. The value of an integer cannot be changed after it is created.
Float (float): Floats are immutable. The value of a float cannot be changed after it is created.
Boolean (bool): Booleans are immutable. Their values are True or False, and they cannot be modified.
Note:

Immutability provides certain advantages, such as hashability (useful in sets and dictionary keys)
and safer sharing of objects in concurrent programming.
Tuples can contain mutable objects (e.g., lists),
but the tuple itself remains immutable (the reference to the mutable object doesn't change).
Some built-in types, like frozenset, are explicitly designed to be immutable sets.
User-defined classes can be designed as mutable or immutable based on how they are implemented.
```
