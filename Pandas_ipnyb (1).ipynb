```python
import pandas as pd
production=pd.Series([30,60,45,25],
                     index=["wheat","rice","maize","barley"])
print(production)
```

```python
#by crop name[label]
print(production["rice"])
print(production.loc["rice"])
```

```python
#by positioning
print(production.iloc[0])
print(production.iloc[2])
```

```python
print(production.mean())
print(production.sum())
print(production.max())
print(production.idxmax())
```

```python
#increase all crop production by 10%
incresed_production=production*1.10
print(incresed_production)
```

```python
#crops with production greater than 40 tons
high_production=production[production>40]
print(high_production)
print()

print(production>40)
```

```python
import pandas as pd
data={"farm":["farm A","farm B","farm C","farm D"],
   "crop":["wheat","rice","maize","barley"],
      "area_hectares":[10,15,8,12],
      "production_tons":[30,60,24,36]}
df=pd.DataFrame(data)
print(df)
```

```python
df.head(2) #first few rows
```

```python
df.tail(2) # last bottom rows
```

```python
df.info() #get basic info
```

```python
df.describe() #summary statistics
```

```python
# add a new coloum(yield)
#yeild= production/yield
df["yield_per_ton_hactare"]=df["production_tons"]/df["area_hectares"]
print(df)
```

```python
print(df.iloc[[1,3]])
```

```python
df.iloc[[1,3]]
```

```python
df.iloc[[0,3,2],[1,2,3]]
```

```python
print(df.loc[[1,2],["farm","crop"]])
```

```python
print(df.loc[0:3])
```

```python
#filter farms growing wheat
print(df["crop"]=="wheat")
print()
wheat_farms=df[df["crop"]=="wheat"]
print(wheat_farms)
```

```python
#farms with yield greater than 3
high_yield=df[df["yield_per_ton_hactare"]>3]
print(high_yield)
```

```python
#sort farms by highest production
sorted_df=df.sort_values(by="production_tons",ascending=False)
print(sorted_df)
```

```python
avg_yield=df.groupby("crop")["yield_per_ton_hactare"].sum()
print(avg_yield)
```
