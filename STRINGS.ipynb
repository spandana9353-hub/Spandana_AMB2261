```python
crops=0
points=0
print("☘️welcome to farm game☘️")
print("1.🌱plants to seed🌱")
print("2.watering💦🚿🌧️")
print("3.harvesting🌾")
print("4.exit🔚")

while True:
   choice=int(input("enter your choice"))
   if choice==1:
    crops+=1
    print("🌱you have planted🌱",crops,"crops")
   elif choice==2:
    if crops >= 1:
     print("you have watered💦🚿🌧️",crops,"crops")
    else:
     print("No crops to water💦🌧️.")
   elif choice==3:
    if crops >= 1:
      points+=crops*10
      print("plants harvested🌾",crops,"crops")
      crops = 0
    else:
      print("No crops to harvest🌾.")
   elif choice==4:
    print("game over🔚! Total points:", points)
    break
   else:
    print("invalid choice")
```

```python
import random
inventory={}
total_money=0
fish_prices={"Rohu":12,"Catla":18,"Mrigal":25}
def showmenu():
 print("\n---fish gaming menu---")
 print("1.catch a fish")
 print("2.view inventory")
 print("3.sell fish")
 print("4.exit")
def catch_fish():
  fish=random.choice(list(fish_prices.keys()))
  inventory[fish]=inventory.get(fish,0)+1
  print(f"you caught a {fish}")
def list_inventory():
  if not inventory:
       print("no fish")
  else:
       print("your inventory:")
       for fish,count in inventory.items():
         print(f"{fish}:{count}")
def view_inventory():
  global total_money
  if not inventory:
       print("no fish to sell")
  else:
      earnings_from_current_sale=0
      for fish,count in inventory.items():
         earnings_from_current_sale+=fish_prices[fish]*count
      total_money+=earnings_from_current_sale
      inventory.clear()
      print(f"you sold all fish for ${earnings_from_current_sale}")
showmenu()
while True:
  choice=int(input("enter your choice:"))
  if choice==1:
    catch_fish()

  elif choice==2:
    list_inventory()

  elif choice==3:
    view_inventory()

  elif choice==4:
    print(f"final money earned:${total_money}")
    print("thank you for playing")
    break
  else:
    print("invalid choice.try again")
```
