```python
#GROUP BY THE YEAR AND SUM PRODUCTION
yearly_production=df.groupby("Year")["Production_Tons"].sum()
print("\nyearly total production")
print(yearly_production)
```

```python
#SCATTER PLOT -RAINFALL VS PRODUCTION
import matplotlib.pyplot as plt
plt.scatter(df["Rainfall_mm"],df["Production_Tons"])
plt.xlabel("rainfall(mm)")
plt.ylabel("production(tons)")
plt.title("Rainfall vs Production")
plt.show()
```

```python
#CREATE BAR PLOT
plt.figure(figsize=(5,4))
plt.bar(df["Year"],df["Production_Tons"],color="green")
plt.xlabel("year")
plt.ylabel("yearly_production(tons)")
plt.title("total crop Production per year")

plt.show()
```

```python
# bar chart- area vs year
plt.figure(figsize=(5,4))
year_grp=df.groupby("Year")
plt.bar(df["Year"],df["Area_Hectares"],color="orange")
plt.xlabel("Year")
plt.ylabel("Area(Hectares)")
plt.title("total cultivated area per year")
plt.xticks(df["Year"])
plt.show()
```

```python
wheat=df[df["Crop"]=="Wheat"]
rice=df[df["Crop"]=="Rice"]
plt.figure()
plt.plot(wheat["Year"],wheat["Production_Tons"],label="wheat",color="r")
plt.plot(rice["Year"],rice["Production_Tons"],label="rice",color="g")
plt.xlabel("year")
plt.ylabel("production(tons)")
plt.title("production of wheat and rice")
plt.legend()
plt.show()
```

```python
#bar plot-production by crop(single year 2022)
data_2022=df[df["Year"]==2022]
print(data_2022)
plt.figure(figsize=(6,4))
plt.bar(data_2022["Crop"],data_2022["Production_Tons"],color=["green","orange"])
plt.xlabel("crop")
plt.ylabel("production(tons)")
plt.title("production of crops in 2022")
plt.show()
```
