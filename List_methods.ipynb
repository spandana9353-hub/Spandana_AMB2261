```python
# = vs copy
a=[12,24,36]
b=a
print("a",a)
print("b",b)
b.insert(1,55)
print("a:",a)
print("b:",b)
```

```python
a=[12,24,36]
b=a.copy()
print("a",a)
print("b",b)
b.insert(1,55)
print("a:",a)
print("b:",b)
```

```python
c=[12,24,36]
d=c.copy()
print("c",c)
print("d",d)
d.insert(1,55)
print("c:",c)
print("d:",d)
```
